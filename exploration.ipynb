```python
import json
import gzip as gz
```

```python
data = None
with gz.open("data/AllPrintings.json.gz") as f:
    for entry in f:
        data = json.loads(entry)

print(data.keys())
```

```python
mtg_sets = data.get('data')
print(mtg_sets.get('ONE').keys())
```

```python
multiverse_ids = set()
id_types = set()
nan_counter = 0
for card_set in mtg_sets.values():
    for card in card_set.get('cards'):
        new_id = card.get('identifiers').get('multiverseId')
        if new_id is None:
            print(f"set: {card_set.get('name')}, card: {card.get('name')}")
            nan_counter += 1        
        multiverse_ids.add(new_id)
        id_types.add(type(new_id))
```

```python
print(len(multiverse_ids))
print(len(id_types))
print(id_types)
print(nan_counter)
```
