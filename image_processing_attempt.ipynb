```python
import torch
import torchvision.transforms as transforms
from PIL import Image
import random
from IPython.display import display
```

```python
# Open a magic card image
img = Image.open("data/images/907.jpeg")
display(img)
```

```python
# Define the rotation angle range and the background color
angle_range = (-60, 60)
background_color = (random.randint(50, 150), random.randint(30, 70), random.randint(10, 30))

# Create a background image with a random shade of brown color
background = Image.new("RGB", (img.size[0], img.size[1]), background_color)
display(background)
```

```python
# Define the transformations
transform = transforms.Compose([
    transforms.RandomRotation(angle_range, expand=True, fill=background_color),
    transforms.ToTensor(),
])

# Apply the transformations to the image
img_transformed = transform(img)
img_transformed
```

```python
img_transformed = transforms.ToPILImage()(img_transformed)
display(img_transformed)
```

```python
img_transformed.size
```

```python
background = background.resize(img_transformed.size)
background.size
```

```python
# Paste the transformed image onto the background image
x = random.randint(0, background.size[0] - img_transformed.size[0])
x_offset = x + img.size[0]
y = random.randint(0, background.size[1] - img_transformed.size[1])
y_offset = y + img.size[1]
print(x, x_offset, y, y_offset)
background.paste(img_transformed, (x, y))
display(background)
```
